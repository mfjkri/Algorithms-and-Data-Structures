```python
import os
import csv

from typing import List, Tuple

import networkx as nx
import matplotlib.pyplot as plt
```

```python
def load_vertice(network_number: int) -> List[Tuple[int, ...]]:

    edges = []

    with open(os.path.join("networks", f"network{str(network_number)}.csv"), 'r') as csv_file:
        myReader = csv.reader(csv_file)
        for row in myReader:
            u, v = row[0], row[1]
            edges.append((u, v))

    return edges
```

```python
edges = load_vertice(network_number=19)
```

```python
# To create an empty undirected graph
G = nx.Graph()
  
G.add_edges_from(edges)
plt.figure(5,figsize=(24,24)) 

nx.draw_networkx(G, font_size=50)
```
